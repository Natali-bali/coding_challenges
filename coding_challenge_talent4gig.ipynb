```python
import pandas as pd
import numpy as np
import folium
import matplotlib.pyplot as plt
import json
```

```python
df_listings = pd.read_csv('/content/listings.csv')
```

```python
df_listings
```

```python
room_types = df_listings['room_type'].unique()
```

```python
room_types
```

```python
# I shorten listings database for time consuming reason
df = df_listings[0:1000]
```

```python
def style_fcn(listing):
    color = 'red'
    listing_room_type = listing['room_type']
    if listing_room_type == 'Private room':
      color = 'green'
    elif listing_room_type == 'Shared room':
      color = 'blue'
    elif listing_room_type == 'Hotel room':
      color = 'yellow'
        
    return color
```

```python
map = folium.Map(location = [45.45139383429476, 9.195926613375928], zoom_start=12, tiles='OpenStreetMap')
```

```python
fg = folium.FeatureGroup(name = 'Milano Listing')
```

```python
html = """
<h3 style = 'color:blue'>Host Name: {}</h3>
<h3 style = 'color: blue'>Listing ID: {},/h3>
<h3>Name: {}</h3>
<h3>Area: {}</h3>
<h3>Price: {} per night</h3>
"""
```

```python
for index, listing in df.iterrows():
  fg.add_child(folium.CircleMarker([listing['latitude'], listing['longitude']], radius = 2,
                             popup = html.format(listing['host_name'], listing['id'], listing['name'], listing['neighbourhood'], listing['price']),
                             color = style_fcn(listing)
                             ))
```

```python
map.add_child(fg)
```

```python
file = open('/content/neighbourhoods.geojson', 'r', encoding='utf-8-sig')
json_file = file.read()
file.close()
json_file = json.loads(json_file)
```

```python
fg.add_child(folium.GeoJson(json_file))
```
