```python
df['days_occupied'] = df['availability_365'].apply(lambda x: 365 - x)
df['income_per_year'] = df['days_occupied']*df['price']
df['income_per_month'] = round(df['income_per_year']/12)
```

```python
def style_fcn_point(listing):
    color = 'IndianRed'
    listing_room_type = listing['room_type']
    if listing_room_type == 'Private room':
      color = 'LightGreen'
    elif listing_room_type == 'Shared room':
      color = 'CornflowerBlue'
    elif listing_room_type == 'Hotel room':
      color = 'yellow'
        
    return color
```

```python
map = folium.Map(location = [45.45139383429476, 9.195926613375928], zoom_start=12, tiles='OpenStreetMap')
```

```python
fg = folium.FeatureGroup(name = 'Milano Listings')
```

```python
html = """
<div width = 150>
    <li style = 'list-style-type: none'>
        <ul>
            <p style = 'color: CornflowerBlue;'>Host Name: {}</p> 
        </ul>
        <ul>
            <p style = 'color: CornflowerBlue; list-style-type: none'>Listing ID: {}</p>
            <p style = 'color: CornflowerBlue'>{}</p>
            <p style = 'font-weight: bold;'>Area: {}</p>
            <p>Room type: {}</p>
        </ul>
        
        <ul>
            <p style = 'font-weight: bold;'>{} euro income monthly</p>
            <p style = 'color:grey'>{} euro per night</p>
            <p style = 'color:grey'>{} nights minimum stay</p>
        </ul>  
        
        <ul>
            <p style = 'font-weight: bold;'>{} nights booked annually</h4>
        </ul>  
    </li>
</div>
"""
```

I will use json area map to create the map of the most expensive area

```python
t = df.groupby('neighbourhood').mean()
```

```python
t
```

```python
t['price'] = round(t['price'])
t.price.max(), t.price.min(), t.price.median()
```

```python
file = open('data/neighbourhoods.geojson', 'r', encoding='utf-8-sig')
json_file = file.read()
file.close()
json_file = json.loads(json_file)
```

```python
def style_fcn_geolocation(geo_loc, df = t):
    color = 'green' 
    price = 0
    location = geo_loc['properties']['neighbourhood']
    locations = list(df.index)
    if location in locations:
        price = df.at[location, 'price'] 
        
        if 60 > price <= 100:
             color = 'yellow'
        elif 100 > price <= 150:
             color = 'orange'
        elif price > 150:
             color = 'red'
    return {"fillOpacity": 0.1,
            "fillColor": color}
```

```python
fg.add_child(folium.GeoJson(json_file, style_function = style_fcn_geolocation))
```

```python
for index, listing in df.iterrows():
  fg.add_child(folium.CircleMarker([listing['latitude'], listing['longitude']], radius = 2,
                             tooltip = html.format(listing['host_name'], 
                                                   listing['id'], 
                                                   listing['name'], 
                                                   listing['neighbourhood'], 
                                                   listing['room_type'],
                                                   listing['income_per_month'],
                                                    
                                                   listing['price'], 
                                                   listing['minimum_nights'],
                                                   listing['days_occupied']),
                             color = style_fcn_point(listing)
                             ))
```

```python
map.add_child(fg)
```

```python
map.save('map.html')
```
